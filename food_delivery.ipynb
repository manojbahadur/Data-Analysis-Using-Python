```python
menu = {0:"Pizza", 1:"Fries", 2:"Chicken Burger", 3:"Veg Burger", 4:"Noodels", 5:"chicken strips", 
        6:"Fried rice", 7:"Lemon Soda", 8:"Coca-Cola", 9:"Tea", 10:"Coffee"}
prices = [99.99,35.45,99.89,69.89,45.80,75.99,60.69,25.58,39.99,4.99,9.99]

def menu_order():
    print("*" * 28 + "ONLINE FOOD ORDERING SYSTEM" + "*" * 24 + "\n") 
    print("*" * 34 + "MAIN MENU" + "*" * 34 + "\n" + "_" * 72)
    print("*" * 33 + "ORDER FOODS" + "*" * 33)
    print(" |NO| |FOOD NAME|         |PRICE|")
    i = 0
    while i < len(menu):
        var_space = 1
        if i <= 8:                      
            var_space = 2
        food = " (" + str(i + 1) + ")" + " " * var_space + str(menu[i])
        price = " " * (19-len(menu[i])) + "$" + " " + str(prices[i])
        print(food,price)
        i += 1
        
def get_order():
    current_order = []
    while True:
        print("What can I get for you?")
        order = int(input())
        if order-1 in menu:
            current_order.append(order-1)
        else:
            print("I'm sorry, we don't serve that.")
            continue
        if is_order_complete():
            return current_order

def is_order_complete():
    print("Anything else? yes/no")
    choice = input()
    if choice == "no":
        return True
    elif choice == "yes":
        return False
    else:
        raise Exception("invalid input")
    
def payment(order):
    print("*" * 32 + "PAYMENT" + "*" * 33 + "\n") 
    total_price = 0.0
    print("Your orders: ")
    for i in order:
        total_price += prices[i]
        print(menu[i],"  $",prices[i])
    print("_" * 72)    
    print("Total: ","$",round(total_price,2))
    tax= round(0.125*total_price,2)
    print("Tax is 12.5%: ","$",tax)
    total = round(total_price+tax,2)
    print("After Tax Total Amount: ","$",total)
    print("_" * 72)
    print("*"*15,"Your order will be deliveredin 30min","*"*15)
    print("*"*15,"Thank You For Ordering!..........:)","*"*15)

menu_order()
order = get_order()
payment(order)
```
